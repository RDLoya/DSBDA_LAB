```python
#IOR
q1=df4['Total'].quantile(0.25)
q2=df4['Total'].quantile(0.50)
q3=df4['Total'].quantile(0.75)
IQR=q3-q1
UL=q3+(1.5*IQR)
LL=q1-(1.5*IQR)
out= ~df4['Total'].between(LL,UL)
df4[out]
```

```python
dfx = df4

q1=dfx['Percentage'].quantile(0.10)
q2=dfx['Percentage'].quantile(0.90)

dfx['mew_Agg']=np.where(dfx["Percentage"]>q2,q2,
                       np.where(dfx["Percentage"]<q1,q1,df4["Percentage"]))

dfx1= dfx.loc[:, dfx.columns!='Name']
print(dfx1)

#Min Max Scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_columns = dfx1.columns
dfx_scaled = pd.DataFrame(scaler.fit_transform(dfx1), columns=X_columns)
dfx_scaled
```
