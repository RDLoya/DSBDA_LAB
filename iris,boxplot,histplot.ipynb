```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
fig, axs = plt.subplots(2,2,figsize=(7,9))
sns.histplot(data = df, x="SepalLengthCm", bins=10, color="skyblue", ax=axs[0,0], kde=True)
sns.histplot(data = df, x="SepalWidthCm", bins=10, color="Teal", ax=axs[0,1], kde=True)
sns.histplot(data = df, x="PetalLengthCm", bins=10, color="olive", ax=axs[1,0], kde=True)
sns.histplot(data = df, x="PetalWidthCm", bins=10, color="green", ax=axs[1,1], kde=True)
```

```python
fig,axs = plt.subplots(2,2,figsize=(9,10))
sns.boxplot(data=df,x="Species",y="SepalLengthCm",color="skyblue",ax=axs[0,0])
sns.boxplot(data=df,x="Species",y="SepalWidthCm",color="Teal",ax=axs[0,1])
sns.boxplot(data=df,x="Species",y="PetalLengthCm",color="olive",ax=axs[1,0])
sns.boxplot(data=df,x="Species",y="PetalWidthCm",color="orange",ax=axs[1,1])
```
